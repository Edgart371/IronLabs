Stationarity is a key part of time series analysis. Import the augmented Dickey-Fuller test from the statsmodels package.

```python
from statsmodels.tsa.stattools import adfuller
```

Let’s pass our data frame into the adfuller method. 
Here, we specify the autolag parameter as “AIC”, which means that the lag is chosen to minimize the information criterion

```python
adft = adfuller(data, autolag='AIC')
print('ADF Test Statistic: %.2f' % adft[0])
print('5%% Critical Value: %.2f' % adft[4]['5%'])
print('p-value: %.2f' % adft[1])
```

Store our results in a dataframe display it

```python
result = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]  , "Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"]})
result
```

Explain the results

```python
#We can see that our data is not stationary from the fact that our p-value is greater than 5 percent
#and the test statistic is greater than the critical value.
```

Autocorrelation
Checking for autocorrelation in time series data is another important part of the analytic process. 
This is a measure of how correlated time series data is at a given point in time with past values, 
which has huge implications across many industries. For example, if our passenger data has strong autocorrelation, we can assume that high passenger numbers today suggest a strong likelihood that they will be high tomorrow as well.
Please calculate and show the autocorrelation

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data)
plt.show()
```

Calculate partial autocorrelation

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(data)
plt.show()
```

Let's check the models on our dataset. Please note, if the dataset is non-stationary, you need to make it stationary

```python
# Transform the data in stacionary by the diference method

data['Difference'] = data['Passengers'].diff()
```

```python
data.head()
```

```python
sns.lineplot(x=data.index, y=data.Difference)
plt.title('Difference')
plt.show()
```

```python
# Check it is now stacionary by aplying Aduller model

result = adfuller(data.Difference.dropna())

print('ADF Test Statistic: %.2f' % result[0])
print('5%% Critical Value: %.2f' % result[4]['5%'])
print('p-value: %.2f' % result[1])
```

```python
# Ho: data is not stationary
# Ha: data is stationary
#Still not stacionary p-value is equal to 0.05
#If P vlaue < 0.05 you reject H0 and conclude that the time series is stationary.
```

```python
# Check it is now stacionary by aplying kpss test

from statsmodels.tsa.stattools import kpss

result = kpss(data.Difference.dropna())
print(result)
print('KPSS Test Statistic: %.2f' % result[0])
print('5%% Critical Value: %.2f' % result[3]['5%'])
print('p-value: %.2f' % result[1])
```

```python
# Ho: data is not stationary
# Ha: data is stationary
#Still not stacionary p-value is equal to 0.10
#If P vlaue < 0.05 you reject H0 and conclude that the time series is stationary.
```

```python
# Transform the data in stacionary by the log method
import numpy as np

data['log'] = np.log(data.Passengers)
data['log'] = data['log'].diff()

data = data.drop(data.index[0])
```

```python
data.head()
```

```python
sns.lineplot(x=data.index, y=data.log)
plt.title('Log Difference')
plt.show()
```

```python
# Check it is now stacionary by aplying Aduller model

result = adfuller(data.log.dropna())

print('ADF Test Statistic: %.2f' % result[0])
print('5%% Critical Value: %.2f' % result[4]['5%'])
print('p-value: %.2f' % result[1])
```

```python
# Ho: data is not stationary
# Ha: data is stationary
#Still not stacionary p-value is equal to 0.07
#If P vlaue < 0.05 you reject H0 and conclude that the time series is stationary.
```

```python
# Check it is now stacionary by aplying kpss test

from statsmodels.tsa.stattools import kpss

result = kpss(data.log.dropna())
print(result)
print('KPSS Test Statistic: %.2f' % result[0])
print('5%% Critical Value: %.2f' % result[3]['5%'])
print('p-value: %.2f' % result[1])
```

```python
# Ho: data is not stationary
# Ha: data is stationary
#Still not stacionary p-value is equal to 0.10
#If P vlaue < 0.05 you reject H0 and conclude that the time series is stationary.
```

```python
# Transform the data in stacionary by the boxcox method
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox.html

from scipy.stats import boxcox

data['boxcox'],parameters= boxcox(data['Passengers'])
data['boxcox']=data['boxcox'].diff()

```

```python
data.head()
```

```python
sns.lineplot(x=data.index, y=data.boxcox)
plt.title('Boxcox Difference')
plt.show()
```

```python
# Check it is now stacionary by aplying Aduller model

result = adfuller(data.boxcox.dropna())

print('ADF Test Statistic: %.2f' % result[0])
print('5%% Critical Value: %.2f' % result[4]['5%'])
print('p-value: %.2f' % result[1])
```

```python
# Ho: data is not stationary
# Ha: data is stationary
#Still not stacionary p-value is equal to 0.01
#If P vlaue < 0.05 you reject H0 and conclude that the time series is stationary.
#It is stationary
```
