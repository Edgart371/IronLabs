#### Find the minimum, mean and maximum of all numeric columns for each company department.



```python
# your code here
employees.groupby('Department')['Salary','Years'].agg(['mean', 'max', 'min'])
```

```python
# your code here
employees.groupby('Department')['Salary'].agg( [lambda x: x.max() - x.min()])
```

#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated.

```python
# your code here
sp = orders[orders['Country']=='Spain'].groupby('CustomerID').agg(['count'])

sp.count()
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here

customers = orders.groupby('CustomerID')['Quantity'].agg(['sum'])

customers.head()

customers[customers['sum'] > 50]
```
