#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
#https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xscale.html
fig, [ax1, ax2] = plt.subplots(1,2)

ax1.set_title('PLOT')

ax1.plot(x,x**2)

ax1.set_xlabel('x label') # Labels
ax1.set_ylabel('y label')

ax2.set_title('PLOT2')

ax2.plot(x,np.exp(x))

ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

ax1.set_xscale("linear")
ax2.set_yscale("log")

plt.show()

```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here

fig, [ax1, ax2] = plt.subplots(1,2)

ax1.set_title('PLOT')

ax1.plot(x,x**2)

ax1.set_xlabel('x label') # Labels
ax1.set_ylabel('y label')
ax1.legend("E")

ax2.set_title('PLOT2')

ax2.plot(x,np.exp(x))

ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.legend("E")

ax1.set_xscale("linear")
ax2.set_yscale("log")

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fitbit.columns
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fitbit_agg1 = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()

fitbit_agg1


# Create a figure of a fixed size and axes
f, ax = plt.subplots(1,2, figsize = (12,4))

# Set the current axes instance to ax[0] - Now plt methods will affect ax[0]
plt.sca(ax[0])

# Plot the weekday steps in the current axes
fitbit_agg1.plot(kind = 'line',linewidth=2)

# Add labels, title, etc
plt.ylabel('Median number of steps')
plt.xlabel('Months')
plt.title('Median Steps walked by month')
plt.xticks(range(13), ['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

# Save the figure
plt.savefig('steps_months.png',bbox_inches='tight')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fitbit_agg2 = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()

fitbit_agg2

# Set the current axes instance to ax[1] - Now plt methods will affect ax[1]
plt.sca(ax[1])

# Plot the weekday steps in the current axes
fitbit_agg2.plot(kind = 'bar',linewidth=2)

# Add labels, title, etc
plt.ylabel('Median number of steps')
plt.xlabel('Work days or Weekend days')
plt.title('Median Steps walked by days')
plt.xticks(range(2), ['Weekend','Weekday'])

# Save the figure
plt.savefig('steps_Work_or_Weekend.png',bbox_inches='tight')

# your code here
plt.show()


```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
        
# Create a df with the columns we are interested in
cols = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary','Steps']
df = fitbit[cols]

# Create a figure of a fixed size and axes
f, ax = plt.subplots(1, 3, figsize = (20,8))

# Iterate to draw each scatter plot
for i in range(3):
    ax[i].scatter(df[cols[i]], df[cols[3]], color = 'darkblue')
    ax[i].set_xlabel(df[cols[i]].name)
    ax[i].set_ylabel('Steps')
```
