We break down the main problem into several sub problems:

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
#Another code

#orders = orders[['CustomerID','amount_spent','Country']]
#orders1 = orders.groupby('CustomerID',as_index=False).sum()
#orders1.sort_values(by='amount_spent', ascending = False)

# Sub 1
# your code here
#orders.columns

orders1 = orders.groupby('CustomerID', as_index=False).agg({'amount_spent':'sum'})
orders1.sort_values('amount_spent', ascending=False, inplace=True)

orders1

orders_contry=orders1.join(orders['Country'])

orders_contry
```

```python
# Sub 2
# your code here

vip1= orders_contry.amount_spent.quantile(0.95) #Calculation of the value above the 95th percentile (aka. 0.95 quantile)

print(vip1)

vip = orders_contry.loc[(orders_contry['amount_spent'] >= vip1)]

vip
```

```python
# Sub 3
# your code here

vip2= orders_contry.amount_spent.quantile(0.75) #Calculation of the value above the 75th percentile (aka. 0.75 quantile)

print(vip2)

pref_vip = orders_contry.loc[(orders_contry['amount_spent'] >= vip2) & (orders_contry['amount_spent'] < vip1)]
pref_vip
```

```python
orders_contry['Client_type'] =np.where(orders_contry['amount_spent']>=1661.640, 'Preferred_customer','')
orders_contry['Client_type'] =np.where(orders_contry['amount_spent']>=5840.182, 'VIP',orders_contry['Client_type'])
#orders_contry.head()
orders_contry['Client_type'].value_counts()
```
