#### Question 1: Find the questions and answers of last month.

```python
#Site to convert time "https://www.epochconverter.com"
# your code here
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1650285356, todate=1652877356,min=20)
questions
answers = SITE.fetch('answers', fromdate=1650285356, todate=1652877356,min=20)
answers
```
