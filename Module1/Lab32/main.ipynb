```python
import pandas as pd
import category_encoders as ce
```

```python
df=pd.DataFrame([['1','231','Self-employed','009','High','2','30200'],['2','765','students','005','high','3','12700'],['3','453','Horeca','007','medium','5','89400'],['4','231','self-employed','009','high','2','30200'],['5','892','finance','003','low','3','740000']],
                index=['0','1','2','3','4'],
                columns=['TransactionID','ClientID','Profession','Bank_dep','Risk','Number of credits','Revenue'])

df
```

```python
df['Profession'] = df['Profession'].str.lower()
df
```

```python
df['Risk'] = df['Risk'].str.lower()
df
```

```python
#https://ml.dask.org/modules/generated/dask_ml.preprocessing.DummyEncoder.html
#Dummy encoder
pd.get_dummies(df)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html
#Label Encoder
from sklearn.preprocessing import LabelEncoder

lbl_ebcode = LabelEncoder()
df_encoded = lbl_ebcode.fit_transform(df['Profession'])
df_encoded
```

```python
lbl_ebcode.fit(df['Profession'])
```

```python
df['Profession_labels'] = lbl_ebcode.fit_transform(df['Profession'])
df.head(5)
```

```python
#https://contrib.scikit-learn.org/category_encoders/backward_difference.html
#Category Encoders
#Backward Difference Encoder
import category_encoders as ce

encoder = ce.BackwardDifferenceEncoder()

encoder.fit(df['Risk'])
```

```python
df_encoded = encoder.fit_transform(df['Risk'])
df_encoded
```

```python
encoder.get_feature_names()
```

```python
df.head(5)
```

```python
df.columns
```

```python
#BaseN Encoder

encoder = ce.BaseNEncoder(cols=['TransactionID', 'ClientID', 'Profession', 'Bank_dep', 'Risk', 'Number of credits', 'Revenue', 'Profession_labels'], return_df=True, base=5)
```

```python
df_encoded=encoder.fit_transform(df)

df_encoded
```

```python
#Original Encoder
#Used for ordinal data. Converts non-numerical data into an integer. Part of Category Encoders Libary

encoder= ce.OrdinalEncoder(df['Risk'])
```

```python
df_encoded = encoder.fit_transform(df['Risk'])

df_encoded
```

```python
#One Hot Encoder
#Used for nominal data - data does not to have any order> for each level of categorical feature a new variabel is created containing either 0 or 1

encoder=ce.OneHotEncoder(df['Risk'])
```

```python
df_encoded = encoder.fit_transform(df['Risk'])

df_encoded
```

```python
#Sum coding/Sum encoding
#It is similar to Dummy encoding, but represents the data using 3 values -1,0 and 1. 
#Value -1 appears where there is a position with only 0. 
#Thanks to that there is one less column comapared to One hot encoder

encoder = ce.SumEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Risk'])

df_encoded
```

```python
#Hashing / Hash encoder
#Hashing is the transformation of arbitrary size input in the form of a fixed-size value. 
#It is a one-way process. Hashing has several applications like data retrieval, checking data corruption, and in data encryption also.
#As a result of hushing I obtain similary as in case of dummy encoding a 2 dimension table with 0, 1 values. 
#However, we can choose the number of columns in which data will be represenetd. 
#This allows to represent for example 6 values data with N elemnts.

encoder=ce.HashingEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Risk'])

df_encoded
```

```python
#Binary Encoding
#This is a combination of hashing and one hot encoding. 
#First, the categoritacl data is transformed into numerical using an ordinary encoder, and then the numbers are transformed into binary numbers.

encoder= ce.BinaryEncoder(cols=['Risk'],return_df=True)
```

```python
df_encoded = encoder.fit_transform(df['Risk'])

df_encoded
```

```python
#Target Encoder
#It converts an array of categorical data into a mean value of a numerical data which is dependent 
#from that categorical data. Returns and array of numbers, but requires a target value.

encoder = ce.TargetEncoder(cols='Profession') 
```

```python
df_encoded = encoder.fit_transform(df['Profession'], df['Revenue'])

df_encoded
```

```python
#CatBoost Encoder
encoder = ce.CatBoostEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Profession'], df['Revenue'])

df_encoded
```

```python
#Count Encoder
encoder = ce.CountEncoder(cols=['Risk'])
```

```python
df_encoded = encoder.fit_transform(df['Risk'])

df_encoded
```

```python
#Generalized Linear Mixed Model Encoder
encoder = ce.GLMMEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Profession'], df['Revenue'])

df_encoded
```

```python
#Helmert coding
#With Helmert coding, each level of the variable is compared to "later" levels of the variable. Used for ordinal variables

encoder = ce.HelmertEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Risk'])

df_encoded
```

```python
#James-Stein Encoder, Similar to Target encoder

encoder = ce.JamesSteinEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Profession'], df['Revenue'])

df_encoded
```

```python
#Leave One Out Encoder
#This is very similar to target encoding but excludes the current rowâ€™s target when calculating the mean target for a level to reduce the effect of outliers.

encoder = ce.LeaveOneOutEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Profession'], df['Revenue'])

df_encoded
```

```python
#M-estimate encoder, Similar to target encoder. Simplified version of it.
encoder = ce.MEstimateEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Profession'], df['Revenue'])

df_encoded
```

```python
#Polynominal Encodind
#Similar to Backward Difference Encoder. This type of coding system should be used only with an ordinal variable 
#in which the levels are equally spaced. Examples of such a variable might be income or education

encoder = ce.PolynomialEncoder(cols=["Risk"])
```

```python
df_encoded = encoder.fit_transform(df, verbose=1)

df_encoded
```

```python
#Weight of Evidence
#WoE is a commonly used target-based encoder in credit scoring.

encoder = ce.WOEEncoder()
```

```python
df_encoded = encoder.fit_transform(df['Profession'], df['Profession_labels'])

df_encoded
```
