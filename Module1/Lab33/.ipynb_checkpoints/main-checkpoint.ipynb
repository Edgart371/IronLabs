```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
# Importing the dataset from Ironhack's database
df=pd.read_excel('/Users/edgartome_1/IronHack/IronLabs/Module1/Lab33/example_data_cleaning.xlsx')
df
```

```python
#Drop the column with department value that is equal for the database and not relevant

df=df.drop(['Department'], axis=1)
df
```

```python
#Lower case of profession
df['Profession']= df['Profession'].str.lower()
df
```

```python
#Fill the missin value with neibother value
df['Profession']= df['Profession'].ffill()
df
```

```python
#Replace student by etudient
df.loc[df['Profession'].str.startswith('student'), 'Profession'] = 'etudient'
df
```

```python
#Replace missing values with mean
df['BirthYear']= df['BirthYear'].fillna('1967')
df
```

```python
df.dtypes
```

```python
df=df.astype({'BirthYear':'int'})
df
```

```python
from sklearn.preprocessing import LabelEncoder
df['Risk_label_encoded'] = LabelEncoder().fit_transform(df.Risk)
df
```

```python
df=df.astype({'TransactionID':'int'})
df
```

```python
df=df.astype({'ClientID':'int'})
df
```

```python
df.loc[df['Risk_label_encoded'] == 0, 'Risk_label_encoded']=3

df
```

```python
def change_year(clientID,changetoyear):
    df.loc[df['ClientID']==clientID, 'BirthYear'] = changetoyear
```
